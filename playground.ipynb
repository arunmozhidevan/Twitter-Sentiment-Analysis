```python
import tweepy # Tweepy - Python library for accessing the Twitter API.
import re # Regular Expression Python module
from textblob import TextBlob # TextBlob - Python library for processing textual data
import os
from dotenv import load_dotenv # load the .env file
load_dotenv()
import pandas as pd
```

```python
twitterApiKey = os.getenv("API_key")
twitterApiSecret = os.getenv("API_secret_key")
twitterApiAccessToken = os.getenv("Access_token")
twitterApiAccessTokenSecret = os.getenv("Access_token_secret")
```

```python
auth = tweepy.OAuthHandler(twitterApiKey,twitterApiSecret)
auth.set_access_token(twitterApiAccessToken,twitterApiAccessTokenSecret)
twitterApi = tweepy.API(auth, wait_on_rate_limit = True)
```

```python
def getTweets(twitterAccount, total_tweets):
    return tweepy.Cursor(twitterApi.user_timeline, 
                        screen_name=twitterAccount, 
                        count=None,
                        since_id=None,
                        max_id=None,
                        trim_user=True,
                        exclude_replies=True,
                        tweet_mode='extended'
                        ).items(total_tweets)
```

```python
tweets_data = []
for i, status in enumerate(tweepy.Cursor(api.search_tweets,label='dev_environment', q= "covid", fromDate=202111010000, toDate = 202111140000, maxResults = 100 ).items(100)):
    try:
        quoted_status = status.quoted_status
        extended_tweet = status.extended_tweet
    except:
        quoted_status = None
        extended_tweet = None
    single_tweet_data = {"status_id": status.id
                         , "user" : status.user.screen_name
                         , "text" : status.text
                         , "extended_tweet" : extended_tweet
                         , "geo" : status.geo
                         , "tweeted_at" : str(status.created_at)
                         , "is_a_retweeted_tweet" : status.retweeted
                         , "is_quote_status" : status.is_quote_status
                         , "retweet_count" : status.retweet_count
                         , "hashtags": status.entities['hashtags']
                         , "user_mentions" : status.entities['user_mentions']
                         , "users_num_followers" : status.user.followers_count
                         , "screen_name_replied_to" : status.in_reply_to_screen_name
                         , "status_replied_to_id": status.in_reply_to_status_id
                         , "tweet_place" : status.place
                         , "lang" : status.lang
                        }
    tweets_data.append(single_tweet_data)

tweets_df = pd.DataFrame(tweets_data)
```
